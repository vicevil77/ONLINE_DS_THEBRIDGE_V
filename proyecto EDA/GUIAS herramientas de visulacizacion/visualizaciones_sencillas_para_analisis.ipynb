<h1 id="Visualizaciones"> Visualizaciones con Matplotlib y Seaborn</h1>

`Matplotlib` y `Seaborn` son dos bibliotecas de visualización de datos en Python que comparten una relación cercana, pero tienen algunas diferencias clave. Aquí hay un resumen de las diferencias entre ellas:

1. **Estilo y apariencia:**
   - **Matplotlib:** Requiere más configuración para obtener visualizaciones estéticamente agradables. Puedes personalizar cada detalle, pero a veces puede llevar más tiempo.
   - **Seaborn:** Viene con estilos predefinidos que hacen que las visualizaciones sean más atractivas de manera predeterminada. Facilita la creación de gráficos elegantes con menos configuración.

2. **Sintaxis:**
   - **Matplotlib:** Ofrece una interfaz de bajo nivel y más manual, donde debes especificar detalles como colores y estilos directamente.
   - **Seaborn:** Proporciona una interfaz de alto nivel que simplifica la creación de gráficos complejos. Tiene funciones específicas para tipos de gráficos comunes y utiliza automáticamente paletas de colores.

3. **Tipos de gráficos:**
   - **Matplotlib:** Es una biblioteca más general y versátil que puede crear una amplia variedad de gráficos. Puedes personalizar casi todos los aspectos de los gráficos.
   - **Seaborn:** Está diseñado específicamente para estadísticas y visualizaciones de datos. Se especializa en tipos de gráficos como histogramas, diagramas de dispersión y mapas de calor.

4. **Paletas de colores:**
   - **Matplotlib:** Ofrece una variedad de mapas de colores, pero la elección y configuración a veces pueden ser más manuales.
   - **Seaborn:** Tiene paletas de colores más estéticas y predefinidas. Facilita la asignación de colores a datos categóricos.

5. **Facilidad de uso:**
   - **Matplotlib:** Es más flexible pero puede requerir más líneas de código para realizar ciertas tareas.
   - **Seaborn:** Simplifica la creación de gráficos comunes, lo que hace que sea más fácil y rápido crear visualizaciones con menos código.

**En resumen**, Matplotlib es una biblioteca fundamental y poderosa que proporciona un control completo sobre la creación de gráficos, mientras que Seaborn es una extensión de Matplotlib que simplifica la creación de visualizaciones estadísticas y mejora la estética por defecto. La elección entre ellas dependerá de tus necesidades específicas y preferencias de estilo.

<h2 id="histogramas">1. Histogramas con Matplotlib y Seaborn:</h2>

```python
# Creamos un DataFrame de ejemplo
data = {'edad': [25, 30, 35, 40, 45, 50, 55, 60, 65, 70]}
df = pd.DataFrame(data)
```

<style>

<!-- Matplotlib -->
<h3 id="hist-matplotlib">1.1 Matplotlib:</h3>

```python
# Histograma con Matplotlib
plt.figure(figsize=(8, 4))
plt.hist(df['edad'], color='#00C6C2')
plt.title('Histograma de Edades (Matplotlib)')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()

# ¿Crees que lo puedes hacer mejor? Juega con los parametros ;)
```

<!-- Diagramas de Barras con Matplotlib y Seaborn -->
<h2 id="diagramas-de-barras">3. Diagramas de Barras con Matplotlib y Seaborn:</h2>

<!-- Matplotlib -->
<h3 id="bar-matplotlib">3.1 Matplotlib:</h3>

```python
# Diagrama de barras con Matplotlib
plt.figure(figsize=(8, 4))
plt.bar(df['edad'], df['salario'], color='orange', edgecolor='black')
plt.title('Diagrama de Barras de Salario por Edad (Matplotlib)')
plt.xlabel('Edad')
plt.ylabel('Salario')
plt.show()

# ¿Crees que lo puedes hacer mejor? Juega con los parametros ;)
```

<!-- Seaborn -->
<h3 id="bar-seaborn">3.2 Seaborn:</h3>
