```python
def obtener_estadisticas(df):

    resultado = pd.DataFrame()
    for col in df.columns:
        datos = {}
        datos['porcentaje_cardinalidad'] = round(df[col].nunique() / len(df) * 100, 2)
        datos['Tipo'] = df[col].dtype
        if pd.api.types.is_numeric_dtype(df[col]):
            datos['media'] = round(df[col].mean(), 2)
            datos['moda'] = "No"
            datos['std'] = round(df[col].std(), 2)
            datos['var'] = round(df[col].var(), 2)
            datos['Q1'] = round(df[col].quantile(0.25), 2)
            datos['mediana'] = round(df[col].median(), 2)
            datos['Q3'] = round(df[col].quantile(0.75), 2)
            datos['Categoria'] = 'numerica continua' if df[col].nunique() > 10 else 'numerica discreta'
        else:
            datos['media'] = "No"
            datos['moda'] = df[col].mode().iloc[0] if not df[col].mode().empty else "No"
            datos['std'] = "No"
            datos['var'] = "No"
            datos['Q1'] = "No"
            datos['mediana'] = "No"
            datos['Q3'] = "No"
            datos['Categoria'] = 'categorica ordinal' if df[col].nunique() > 2 else 'categorica nominal'
        datos['porcentaje_NaN'] = round(df[col].isna().mean() * 100, 2)
        resultado[col] = pd.Series(datos)
    return resultado.transpose()
```

```python
# creo tuplas prefijo-identificador

def dividir_prefijo_sufijo(identificador):
    """
    Divide un identificador de característica en su prefijo y su identificador.

    Args:
        identificador: El identificador de característica a dividir.

    Returns:
        Un tupla con el prefijo y el identificador.
    """

    match = re.match(r"^(.*)\.fid--(.*)$", identificador)
    if match:
        return match.groups()
    else:
        return None


```

```python
  # Separar el prefijo y el identificador
df_temp["Pre_fij"] = df_temp["FID"].apply(dividir_prefijo_sufijo)
```

```python
# utilizar el identificador de característica para identificar la marca de tiempo

def identificar_tipo(identificador):
    """
    Identifica el tipo de datos de un identificador de característica.

    Args:
        identificador: El identificador de característica a identificar.

    Returns:
        El tipo de datos de los datos.
    """

    if identificador[0].startswith("temp_anual_espacial_p"):
        return "Temperatura media anual"
    else:
        return None
```

```python
 # Identificar el tipo de datos
df_temp["Tipo"] = df_temp["Pre_fij"].apply(identificar_tipo)
```

```python
#conseguir dentificar una marca de tiempo específica

def identificar_marca_tiempo(identificador):
    """
    Identifica la marca de tiempo de un identificador de característica.

    Args:
        identificador: El identificador de característica a identificar.

    Returns:
        La marca de tiempo de los datos.
    """

    return identificador.split(".fid--")[1]
```

```python
def coordenadas(new_dataframe):
  
    def convertir_a_tupla(x):
        if isinstance(x, str) and x.startswith('POINT ('):
            coords = x.replace('POINT (', '').replace(')', '').split()
            return float(coords[0]), float(coords[1])
        else:
            return None

    new_dataframe["coordenadas"] = new_dataframe["geom"].apply(convertir_a_tupla)

    return new_dataframe

```

```python
# Función para convertir un identificador de tiempo
def convertir_identificador_tiempo(identificador_tiempo):
    partes = identificador_tiempo.split('_')

    # Convertir la parte de fecha y hora hexadecimal a decimal (asumiendo milisegundos)
    fecha_hora_hex = partes[1]
    fecha_hora_decimal = int(fecha_hora_hex, 16) / 1000  # Convertir a segundos

    # Convertir la parte de diferencia horaria hexadecimal a decimal
    diferencia_horaria_hex = partes[2]
    diferencia_horaria_decimal = int(diferencia_horaria_hex, 16)

    try:
        # Obtener la fecha y hora en formato legible
        fecha_hora_legible = datetime.datetime.utcfromtimestamp(fecha_hora_decimal).strftime('%Y-%m-%d %H:%M:%S')

        # Aplicar la diferencia horaria
        fecha_hora_con_diferencia = datetime.datetime.utcfromtimestamp(fecha_hora_decimal + diferencia_horaria_decimal).strftime('%Y-%m-%d %H:%M:%S')

        return pd.Series({'Fecha_hora_original': fecha_hora_legible, 'Fecha_hora_dif_hor': fecha_hora_con_diferencia})

    except OSError as e:
        return pd.Series({'Fecha_hora_original': 'Error al convertir', 'Fecha_hora_dif_hor': 'Error al convertir'})

# Aplicar la función a la columna "Marca_time"
df_modificado_temp[['Fecha_hora_original', 'Fecha_hora_dif_hor']] = df_modificado_temp['Marca_time'].apply(convertir_identificador_tiempo)
```

```python
# Separar la columna "Marca_time" en dos columnas: "Fecha" y "Hora"

# Separar la columna "Fecha_hora_original" en dos columnas: "Fecha_UTC" y "Hora_UTC"
df_modificado_temp[['Fecha_UTC', 'Hora_UTC']] = df_modificado_temp['Fecha_hora_original'].str.split(' ', expand=True)

# Separar la columna "Fecha_hora_dif_hor" en dos columnas: "Fecha_local" y "Hora_local"
df_modificado_temp[['Fecha_local', 'Hora_local']] = df_modificado_temp['Fecha_hora_dif_hor'].str.split(' ', expand=True)
```
