```python
# nombre de los lugares que tiene las coordenadas

def obtener_info_lugar(longitud, latitud):
    geolocalizador = Nominatim(user_agent="tu_aplicacion")

    try:
        ubicacion = geolocalizador.reverse((latitud, longitud), language='es', exactly_one=True)
        print(f"Coordenadas: ({longitud}, {latitud}), Resultado: {ubicacion}")
        if ubicacion is not None:
            return ubicacion.address
        else:
            return "Coordenadas en el mar u océano"
    except Exception as e:
        print(f"Error al obtener información de ubicación para ({longitud}, {latitud}): {e}")
        return None


# Convertir las coordenadas a punto flotante
df_ip_destino['Longitud'] = df_ip_destino['Longitud'].astype(float)
df_ip_destino['Latitud'] = df_ip_destino['Latitud'].astype(float)

# Obtener una lista de información de lugar
info_lugar_lista = []
for index, row in df_ip_destino.iterrows():
    longitud, latitud = row['Longitud'], row['Latitud']
    info_lugar = obtener_info_lugar(longitud, latitud)
    info_lugar_lista.append(info_lugar)

# Mostrar la lista de información de lugar
for lugar in info_lugar_lista:
    print(lugar)

# Guardar la lista en un archivo de texto
archivo_resultados = 'resultados_geolocalizacion_destino_lugares.txt'
with open(archivo_resultados, 'w', encoding='utf-8') as file:
    for lugar in info_lugar_lista:
        file.write(f"{lugar}\n")

print(f"Resultados guardados en '{archivo_resultados}'.")
```

```python
# Lista para almacenar índices de filas que cumplen con la condición
indices_a_eliminar = []

# Iterar sobre las filas del DataFrame
for index, row in columna_destino.iterrows():
    # Verificar la condición en la columna 'Info_Lugares'
    if row['Column1'] == 'Coordenadas en el mar u ocÃ©ano' or row['Column1'] == "none":
        # Agregar el índice a la lista
        indices_a_eliminar.append(index)

# Eliminar las filas con los índices recopilados
columna_destino.drop(indices_a_eliminar, inplace=True)
```

```python
import pandas as pd
import spacy

# Cargar el modelo de lenguaje de spaCy
nlp = spacy.load('es_core_news_sm')


# Función para clasificar entidades
def clasificar_entidades(row):
    if pd.notna(row['Column1']):  # Verificar si el valor no es nulo
        ciudades = []
        paises = []
        
        for entidad in nlp(row['Column1']).ents:
            if entidad.label_ == 'GPE':
                ciudades.append(entidad.text)
            elif entidad.label_ == 'LOC':
                paises.append(entidad.text)
        
        return pd.Series({'Ciudad': ', '.join(ciudades), 'Pais': ', '.join(paises)})
    else:
        return pd.Series({'Ciudad': None, 'Pais': None})

# Aplicar la función a cada fila del DataFrame
resultados = columna_destino.apply(clasificar_entidades, axis=1)

# Combinar los resultados con el DataFrame original
columna_destino = pd.concat([columna_destino, resultados], axis=1)

# Mostrar el DataFrame resultante
print(columna_destino[['Column1', 'Ciudad', 'Pais']])
```

```python
#GEOLOCALIZAMOSPOR PAIS, HABIENDO QUITADO ZONAS DE MAR Y ZONAS NO LOCALIZABLES DE DESTINO

# Filtrar el DataFrame para quedarse solo con las filas que tienen información de geolocalización
df_lugares_filtrado = columna_destino[['Pais']].dropna(subset=['Pais'])

# Convertir la columna 'Pais' a una lista de ubicaciones
lista_origen = df_lugares_filtrado['Pais'].tolist()

# Crear un objeto de mapa centrado en una ubicación predeterminada
mapa = folium.Map(location=[0, 0], zoom_start=2)

# Agregar un grupo de marcadores
marker_cluster = MarkerCluster().add_to(mapa)

# Crear un objeto geocoder de Nominatim
geolocalizador = Nominatim(user_agent="mi_aplicacion")

# Agregar marcadores al grupo con iconos personalizados
for pais in lista_origen:
    # Obtener las coordenadas mediante geocodificación
    ubicacion = geolocalizador.geocode(pais)
    try:
        if ubicacion:
            # Definir la ruta del icono personalizado
            icono_personalizado = folium.CustomIcon(icon_image=r'C:\Users\victo\Pictures\internet_icon.png', icon_size=(30, 30))
            
            # Agregar el marcador con el icono personalizado
            folium.Marker(location=[ubicacion.latitude, ubicacion.longitude], popup=pais).add_to(marker_cluster)
    except Exception as e:
        print(f"Error al geocodificar {pais}: {e}")
        continue
# Guardar el mapa como un archivo HTML
mapa.save('E:\Cursos\BC_Data_Science\Repositorio\ONLINE_DS_THEBRIDGE_V\proyecto EDA\mapa_geolocalizado_ips_destino_final_con_calor.html')

# Mostrar el mapa en la notebook
display(mapa)
```
