```python

# Contar las ocurrencias de cada combinación de 'Ciudades' y 'Estados' en ambos DataFrames
num_df_eda = df_eda.groupby(['Ciudades', 'Estados']).size().reset_index(name='Conteo_eda')# agrupamos por ciudades y estados contando con size por valores unicos, y agrupando el resultado
# en la columna dela variable name
num_df = df.groupby(['Ciudades', 'Estados']).size().reset_index(name='Conteo_df')

# Fusionar los resultados en un nuevo DataFrame
resultado = pd.merge(num_df_eda, num_df, on=['Ciudades', 'Estados'], how='outer').fillna(0)

# Calcular la suma total de ocurrencias
resultado['Total'] = resultado['Conteo_eda'] + resultado['Conteo_df']

# Filtrar las filas donde hay coincidencias (donde Total > 0)
df_conteo_ciudades = resultado[resultado['Total'] > 0]

# Mostrar el resultado
df_conteo_ciudades.head(1)
```
