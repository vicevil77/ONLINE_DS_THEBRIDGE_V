## Pasos del proyecto

1. Elegir temtica
    1. **Elegir una temtica y buscar datos en funcin de la misma**. Puede ocurrir que no encuentres datos que te sirvan para tu propsito, o que tengas que conseguirlos mediante APIs, scrapping, descarga de BBDD.... No obstante, esta opcin es la que ms se asemejara a un caso real, ya que una gran problemtica en este tipo de proyectos es encontrar datos.

    2. **Directamente ir a pginas de datos abiertos y escoger un dataset de tu inters**. Si no tienes clara la temtica puedes buscar en alguno de los enlaces de datos (recomendable Kaggle) y escoger un tema que te motive.

2. Obtencin de datos. Puedes llevar a cabo el proyecto con estos datos?
3. Define tu hiptesis. Qu piensas que puedes analizar de estos datos? Qu vas a poder resolver? Cmo lo vas a llevar a cabo? (**Fecha lmite Viernes 23**)
4. Preprocesado: obtn todos los datos de las diferentes fuentes que hayas utilizado, juntalos y ponlos en un formato entendible.
5. Limpia los datos: transformaciones, duplicados, missings, columnas intiles...
6. Exploratorio: obtn todos los estadsticos y grficos que necesites para entender bien tu dataset y apoyar tus argumentos.
7. Concluye con tu anlisis si estabas o no en lo cierto acerca de tu planteamiento y tu hiptesis.
8. Crea un dashboard para representar tu anlisis exploratorio si lo consideras necesario.
9. Crea una presentacin con un hilo conductor sobre tu anlisis exploratorio
10. Documenta tu proyecto y sbelo a GitHub.
