# SSDEEP (Fuzzy Hashing) y TLSH (Locality Sensitive Hashing)

1. tecnicas para analisis de malware - comparacion y detecccion de archivos maliciosos:
    - SSDEEP:
    Función: Crea una "huella digital" del contenido de un archivo, capturando patrones y características clave.
    Utilidad: Permite comparar archivos con alta precisión, incluso si han sido modificados levemente (por ejemplo, empaquetado, ofuscación).
    Algoritmo: Divide el archivo en bloques y calcula un "hash" para cada bloque. El hash final se compone de una serie de estos hashes de bloque concatenados.
    Ventajas: Eficiente en tiempo y espacio, ideal para grandes conjuntos de datos.
    Desventajas: Menos preciso que TLSH para detectar cambios menores en el contenido del archivo.
    - TLSH:
    Función: Similar a SSDEEP, crea una huella digital del archivo, pero con mayor precisión.
    Utilidad: Detección de variantes de malware con pequeñas modificaciones, como cambios en nombres de funciones o variables.
    Algoritmo: Divide el archivo en "percepciones" (secuencias de bytes) y calcula un "hash" para cada percepción. La huella digital final se compone de una combinación de estos hashes de percepción.
    Ventajas: Mayor precisión que SSDEEP para detectar modificaciones sutiles.
    Desventajas: Más lento y consume más recursos que SSDEEP.
