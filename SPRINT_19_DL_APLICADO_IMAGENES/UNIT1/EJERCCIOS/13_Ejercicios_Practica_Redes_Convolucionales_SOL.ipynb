````python
def read_data(directorio, reshape_dim = (32,32)):
    X = [] 
    y = []
    for folder in os.listdir(directorio):
        print(folder)
        if os.path.isdir('/'.join([directorio, folder])):
            for file in os.listdir('/'.join([directorio, folder])):

                image = imread('/'.join([directorio, folder, file]))
                image = cv2.resize(image, reshape_dim) # Redimensionamos las imgenes a 32x32

                X.append(image)
                y.append(folder)

    return np.array(X),np.array(y)
````

````python
X_train,y_train = read_data(TRAIN_PATH)
X_test, y_test = read_data(TEST_PATH) # Nota: podramos dejar la creacin de X_test para el final antes de usarlo y as ahorra memoria mientras no tengamos que emplearlo
````

````python
def show_images_batch(paisajes, names = [], n_cols = 5, size_scale = 2):
    n_rows = ((len(paisajes) - 1) // n_cols + 1)
    plt.figure(figsize=(n_cols * size_scale, n_rows * 1.1*size_scale))
    for index, paisaje in enumerate(paisajes):
        plt.subplot(n_rows, n_cols, index + 1)
        plt.imshow(paisaje, cmap = "Greys")
        plt.axis("off")
        if len(names):
            plt.title(names[index])
````

### Ejercicio 4

Disea la arquitectura de la red. Recuerda que es un algiritmo de clasificacin. Ojo con las dimensiones de la entrada. Como gua, copia o parte de la arquitectura del workout (un par de capas Conv2D + MaxPooling2D con activacin relu y despus la fully connected layer)


````python
# Primero definimos las capas de la red, en este caso partimos de 30 filtros de 3x3, seguido de una capa de pooling de 2x2,
# El input shape es el tamao de las imgenes de entrada, en nuestro caso 32x32x3. 
# El tamao del kernel es pequeo pero suele ser de este tipo y ms con resoluciones tan bajas, el nmero de filtros est escogido de aquella manera,
# juega con el nmero de filtros.

capas = [keras.layers.Conv2D(filters = 30, kernel_size = (3, 3),
                            input_shape = (X_train[0].shape), 
                            padding = "same",
                            activation = 'relu'),
         keras.layers.MaxPooling2D(pool_size = (2, 2)),

# Aadimos otra capa de convolucin y pooling
         keras.layers.Conv2D(filters = 15, 
                             kernel_size = (3, 3),
                             padding = "same", 
                             activation = 'relu'),
         keras.layers.MaxPooling2D(pool_size = (2, 2)),

# Aadimos una capa de flatten para convertir la salida de la capa de pooling en un vector
         keras.layers.Flatten(),

# Aadimos dos capas densas para la clasificacin 
         keras.layers.Dense(units = 150, activation = 'relu'),
         keras.layers.Dense(units = 100, activation = 'relu'),

# Aadimos la capa de salida con 6 neuronas, una por cada clase
         keras.layers.Dense(6, activation = 'softmax')]

# Creamos el modelo
model = keras.Sequential(capas)
````

### Ejercicio 6

Entrena el modelo reservando un 20% de los datos del entrenamiento para validar y con los callbacks que creas necesarios. NOTA: Ojo, tendrs que preparar un poco los X,y antes.



*Primero tendremos que convertir nuestro target a un ordinal :-)*

````python
mapa = {tipo:indice for indice,tipo in enumerate(targets.unique())} # Rehusando la serie que creamos para ver la distribucin del target
mapa
````

````python
y_train_num = np.array([mapa[y] for y in y_train])
y_test_num = np.array([mapa[y] for y in y_test])
````

*Adems los datos estn ordenados, por lo que debemos hacer un shuffle*

````python
# Revolvemos los datos para que el modelo no aprenda en orden de las imgenes
X_train, y_train_num = shuffle(X_train, y_train_num)
````

````python
history = model.fit(X_train,
                    y_train_num,
                    validation_split = 0.2,
                    epochs = 100,
                    batch_size = 32, callbacks= [keras.callbacks.EarlyStopping(patience = 10, restore_best_weights= True)]) # Queremos recuperar los pesos de la mejor de las epocas durante el periodo de patience
````


### Ejercicio 7
