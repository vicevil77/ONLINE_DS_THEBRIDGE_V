Antes de continuar con el miniEDA y la trasnformacin de features, siempre que trabajemos con imgenes, tener a mano una forma de visualizarlas nos ayudar:

```python
def plot_digits(digits,  labels = [], names = ["cero","uno","dos","tres","cuatro","cinco","seis","siete","ocho","nueve"],n_cols = 5,):
    n_rows = ((len(digits) - 1) // n_cols + 1)
    plt.figure(figsize=(n_cols * 2, n_rows * 2.2))
    if isinstance(digits,pd.DataFrame):
        digits_val = digits.values.reshape(-1,28,28)
    else:
        digits_val = digits
    for index, digit in enumerate(digits_val):
        plt.subplot(n_rows, n_cols, index + 1)
        plt.imshow(digit, cmap = "Greys")
        plt.axis("off")
        if len(labels):
            if isinstance(labels, pd.Series):
                plt.title(names[labels.iloc[index]])
            else:
                plt.title(names[labels[index]])
        
```

### Modelos DL con redes convolucionales (pequeo repaso)

Como todo modelo DL, lo "farragoso" es hacer la seleccin correcta de los primeros hiperparmetros y variables de configuracin (y lo complicado es ajustarlos despus :-)):
1. Topologa: Capas, unidades por capa, funciones de activacin, conexin entre capas.
2. Optimizador: Qu optimizador, learning_rate (valor inicial, constante, adaptativo, etc), qu valores para los hiperparmetros propios del optimizador elegido.
3. (No son hiperparmetros, pero...) Problema: Funcin de prdida, mtrica
4. Entrenamiento: Batch Size, Epochs, Callbacks (Earlystop, Checkpoints,etc)

### Topologa

Adems de ajustar los hiperparmetros que definen la topologa, vamos a ver un nuevo tipo de capa que nos permitir obtener modelos que generalizan mejor. Es decir es un tipo de capa, no entrenable, que regulariza el modelo.

Lo primero es, por tanto, que topologa. Vamos a montar algo similar a lo que nos muestra la siguiente figura:

Vamos a crear dos juegos de capas convolucional y de pooling (con funcin de pooling el Mximo), terminaremos la red con una capa oculta y una capa de salida con 10 unidades (porque hay diez clases) a las que aplicaremos una funcin de activacin "softmax" como ya sabemos que tenemos que hacer con problemas de clasificacin.

Y aadiremos esas nuevas capas de "Dropout" (o poda) que nos servirn para regularizar el modelo (que se colocan cuando el modelo tiene cierta profundidad o cierta anchura, es decir hemos expandido mucho las features).

Ah, y nos queda una capa instrumental que no est dibujada que es la que pasar de la parte convolucional a la densa, es decir una capa "Flatten" para volver a arrays unidimensionales que son los que daremos como entrada la cabeza densa de la red.

Empecemos:

```python
import tensorflow.keras as keras
import tensorflow as tf
# De aqu sacamos el tipo de modelo que vamos a construir, que sigue siendo una secuencia o cadena de capas consecutivas
model = tf.keras.Sequential() # creamos el modelo para ir aadiendo las capas
```

Aadimos la primera **Capa convolucional**:
* 8 **filtros** (es como tener 8 unidades "convolucionales"), cuntos filtros necesito? Depende, tendrs que jugar con este nmero en funcin del tamao de entrada de las imagenes y el detalle fino inicial que quieras (para imagenes ms grandes 16,32,64,... estas son realmente pequeas poco detalle fino puedo encontrar).  
* 5x5 de **tamao de Kernel (o campo receptivo, tambin se denomina as al kernel)**. Lo tpico son 3,5 y 7. El valor depende de lo que queramos hilar de fino respecto a lo que consideremos un grupo de pixeles. Aqu vamos a pensar que cuadrado de 25 pixeles (5x5) es la agrupacin adecuada. Juega con esos tres valores. Pero si la imagen est ms pixelada (ms borrosa) igual te merece usar kernels ms grandes.  
* **Stride** de 1x1, es decir vamos a aplicar el kernel a todos los pxeles de la entrada. Esto suele ser as generalmente con la parte de la entrada  
* **Padding** a "Same", acorde con el stride de 1x1, vamos a rellenar "virtualmente" los bordes de forma que estos entren en el anlisis. Same es la forma de no tener que preocuparse por el tamao del kernel para ajustar el padding, lo autoajusta por dentro.
* Para terminar, como **funcin de activacin** usaremos "ReLU" (que en este tipo de capas es lo habitual).

```python
#from keras.layers import Conv2D # Esta es la capa para convolucionales con entrada de 2 dimensiones 
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense # Esta es la capa para convolucionales con entrada de 3 dimensiones


model.add(Conv2D(filters = 8, kernel_size = (5,5),padding = 'Same', 
                 activation ='relu', input_shape = (28,28,1))) # Fijate en dos cosas, en la el input_shape (ancho,alto,num_canales) 
                                                            # y en que no decimos nada de strides (el argumento para el stride), porque por defecto est a (1,1)

```

Aunque no es fundamental que lo sepas a hacer, vamos a obtener el nmero de pesos que tendr esta capa y la resolucin de los mapas de caractersticas a la salida:

* Pesos = num_mapas (o num filtros) * (tamao del kernel (ancho x alto) + 1 (el bias de cada unidad, recuerda que un filtro es como una "neurona convolucional"))  
* Resolucin: El stride (o strides) es 1x1 y el padding es completo -> La resolucin es igual a la salida es decir cada filtro va a tener 28x28 pxeles (784 features)

O sea:

```python
print("Pesos a entrenar:", 8 * ( 5 * 5 + 1))
print("Resolucin (features):", 28/1 * 28/1, " pero ojo como hay 8 filtros, las features las hemos ampliado a:", 28*28*8)
```

Ahora vamos con la **capa de pooling**, en este caso los hiperparmetros:( reducimos dimensionalidad sin perder informacion)
* Tipo de pooling, el mximo (es lo habitual), lo norm,al es quedarse con el pixel de mayor intensidad
* kernel de 2x2
* strides igual que el kernel 2x2

A la salida de cada cpa de pooling sera la litad 14x14 x8 subpas, capas como capas de la convulation

Qu queremos hacer con esta es reducir la dimensionalidad en concreto bajando la resolucin de los mapas de caractersticas en 2x2 (el tamao del stride) y quedndonos con el mximo de cada agrupacin de 4 pxeles (2x2 del kernel)...

```python
from tensorflow.keras.layers import MaxPool2D # Esta es la clase que corresponde a las capas de Pooling que hacen maxpooling sobre datos de entrada en 2D
model.add(MaxPool2D(pool_size=(2,2), strides = (2,2)))# el pill-size es el kernel del pooling
```

Cuntos pesos tengo que entrenar en esta capa?...

Cul es la "resolucin" de salida y por tanto las features que tendremos (para ver la reduccin)?:
$$\frac{28}{2}\times \frac{28}{2} \times \text{Numero de "filtros" de pooling}$$

Y cuantas capas de pooling si no hemos dicho nada? Las capas de pooling se adaptan a la capa que tengan "delante", en este caso esta capa de pooling tendr 8 filtros o repeticiones porque es el nmero de filtros de la anterior capa (la convolucional). De todo esto se encarga `Keras`

$$\frac{28}{2}\times \frac{28}{2} \times 8 = 6272/4 = 1568 \text{ features}$$

hemmos aplanado con el pooling las features

Que corresponden a la bajada a 14x14 la resolucin de cada mapa de caractersticas ($\frac{28}{2} \times \frac{28}{2}$)

```python
14*14*8
```

### Dropout

Lo que toca ahora es crear esa misteriosa capa de DropOut. El dropout o poda es un tcnica de regularizacin que consiste en seleccionar aleatoriamente unidades o neuronas que se "eliminan" (podan) en el entrenamiento. En cada batch se anulan las entradas a algunas de las neuronas de las capas que vienen tras una capa de dropout y eso consigue que el modelo se esfuerce ms. Es por eso una tcnica de regularizacin.

Las capas de dropout se colocan cuando la red empieza a ser muy profunda o a expandir mucho sus features. Solo tienen un hiperparmetro que es su **ratio** que es el porcentaje de neuronas que se bloquean en cada batch. Suele estar entre 0.2 y 0.5 pero depende del nmero de unidades de la capa que va detrs (si tengo 2 unidades un dropout de ms de 0.5 anula la capa por ejemplo).

Otra caracterrstica importante, **slo funcionan en el entrenamiento, en la inferencia o prediccin las capas de dropout se quitan** (lo hace keras por nosotros). En definitiva lo que hacen es filtrar la entrada para escoger tantas como el ratio indique y ponerlas a cero a su salida que es la que se conecta con la siguiente capa.

Para terminar, **no son entrenables** (no tienen parmetros).

Creemos nuestra capa de dropout con un ratio del 0.25 (es decir vamos a anular el 25% de las salidas de la pareja convolucional-pooling que hemos definido ya):

```python
from tensorflow.keras.layers import Dropout

model.add(Dropout(rate= 0.25))
```

Ahora construimos la siguiente seccin:

```python
model.add(Conv2D(filters = 16, kernel_size = (3,3),padding = 'Same', 
                 activation ='relu'))
# Resolucin a la salida: Igual stride(1,1) -> 14x14
model.add(MaxPool2D(pool_size=(2,2), strides=(2,2)))
# Resolucin a la salida: 14/2 x 14/2 -> 7x7
model.add(Dropout(0.25))

```

Y ahora la "cabeza" de la red donde realmente se va a hacer ya la clasificacin a partir de todas las features que nos ha creado nuestra parte convolucional:

Observa que aadimos primero el aplanamiento de la salida de los filtros de pooling para poder atacar a la densa... Este flatten no es cualquier flatten porque no solo aplana los 7x7 en 49 sino que adems concatena cada mapa o filtro (que ahora son 16), es decir a la salida de la capa Flatten tendremos:


```python
from tensorflow.keras.layers import Flatten, Dense

model.add(Flatten())
print("Entrada de la capa flatten:", model.output.shape)# el output es la ultima salida y se le puede hacer shape para ver su numero
model.add(Flatten()) # 49 features * Nmero de filtros o mapas de caractersticas -> 49*16 = 784 features(28x28) -> Volvemos a las features de entrada, esto es muy tpico
print("Salida de la capa flatten:",model.output.shape)
model.add(Dense(256, activation = "relu"))# el 256 puede cambiarse en potencias de 2, osea 256, 512, 1024, 2048, 4096, 8192, 16384, 32768, 65536, 131072, 262144, 524288, 1048576, 2097152, 4194304, 8388608, 16777216, 335544
# Pesos aqu (49*16+1)*256
model.add(Dropout(0.5))
model.add(Dense(10, activation = "softmax"))
# Pesos aqu (256+1)*10
```

Pues ya tenemos la primera y ms farragosa de las elecciones de hiperparmetros. Tenemos la topologa continuemos con el optimizador, compilemos y entrenemos.

### Optimizador

Para **el optimizador** lo habitual ahora es emplear el **"Adam"** siempre que tengamos un size del dataset de entrenamiento bastante grande:

```python
X_train.size
```

S, tiene ms de un milln de datos, escogemos "Adam".

Ahora el hiperparmetros comn a todos los optimizadores **el leraning rate**: En este caso empezaremos con un buen warm-up (Calentamiento es decir con un learning rate bajo), 0.001

Adems "Adam" es un optimizador que tiene otros hiperparmetros para controlar la forma en que evolucioan algunos de sus clculos con el tiempo son los **hiperparmetros: beta1 y beta2** pero que vamos a dejar a sus valores por defecto (0.9 y 0.999, respectivamente). Puedes ahondar ms en los optimizadores [aqui](https://keras.io/api/optimizers/)

```python
from tensorflow.keras.optimizers import Adam

optimizer = Adam(learning_rate =0.001, beta_1=0.9, beta_2=0.999) #Son los parmetros por defecto, no tendramos que haberlos indicado siquiera.
```

### Compilamos el modelo

Keras y Tensorflow exigen la compilacin del modelo, no son dinmicos y en esa compilacin tenemos que indicarle el optimizador ya creado y la funcin de prdida y la mtricas:

* Optimizador: Adam, ya configurado.
* Loss: sparse_categorical_crossentropy (Por que tenemos como target un escalar si fuera un vector (con un valor por cada clase) tendra que ser "categorical_crossentropy")
* Metrica: "accuracy" (porque el dataset est equilibrado)

```python

# Compile the model
#model.compile(optimizer = optimizer , loss = "categorical_crossentropy", metrics=["accuracy"], run_eagerly= True)
model.compile(optimizer = optimizer , loss = "sparse_categorical_crossentropy", metrics=["accuracy"])
```
