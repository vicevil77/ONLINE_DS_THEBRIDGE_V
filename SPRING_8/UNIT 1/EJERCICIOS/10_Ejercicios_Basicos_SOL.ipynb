### Ejercicio 0

Importa los paquetes y módulos que necesites a lo largo del notebook

```python

import matplotlib.pyplot as plt
import numpy as np
```

### Ejercicio 1

Construye la siguiente figura con dos subplots utilizando el estilo MATLAB (es decir plt.\<funcion\>, no usando objetos `Axes`)

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)


plt.subplot(1, 2, 1)  # 1 fila, 2 columnas, primer subplot
plt.plot(x, y1)
plt.title('Seno')

plt.subplot(1, 2, 2)  # 1 fila, 2 columnas, segundo subplot
plt.plot(x, y2)
plt.title('Coseno')

plt.tight_layout()
plt.show()
```

### Ejercicio 2

Reproduce la imagen de la figura usando tres subplots en una columna, utilizando objetos `Axes`. Las dos primeras son las funciones anteriores y la tercera es la tangente (np.tan)

```python

fig, axs = plt.subplots(3, 1)  # 3 filas, 1 columna

axs[0].plot(x, y1, 'r')  # Rojo
axs[1].plot(x, y2, 'g')  # Verde
axs[2].plot(x, np.tan(x), 'b')  # Azul

plt.tight_layout()
plt.show()
```


### Ejercicio 3: Ajuste de Colores y Estilos de Línea  

Modifica la primera figura para que se muestre de la siguiente manera

```python

fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.plot(x, np.sin(2 * x), 'm--')  # Magenta, línea discontinua
ax2.plot(x, np.cos(2 * x), 'c-.')  # Cian, línea-punto

plt.tight_layout()
plt.show()
```

### Ejercicio 4

Reproduce la figura siguiente. Recuerda que las funciones seno hiperbólico y coseno hiperbólico se obtienen así:
* np.sinh, seno hiperbólico;
* np.cosh, coseno hiperbólico;


```python
plt.subplot(2, 1, 1)
plt.plot(x, np.sinh(x), 'r')
plt.title('Seno Hiperbólico')
plt.xlabel('x')
plt.ylabel('sinh(x)')

plt.subplot(2, 1, 2)
plt.plot(x, np.cosh(x), 'b')
plt.title('Coseno Hiperbólico')
plt.xlabel('x')
plt.ylabel('cosh(x)')

plt.tight_layout()
plt.show()
```

### Ejercicio 5: Ajustando el Espaciado de Subplots

Reproduce la siguiente figura con cuatro subplots, ajusta el espaciado entre los subplots para evitar la superosición (tal y como ves en la figura), emplea `subplots_adjust` y los parámetros hspace y wspace.

```python

fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(x, y1)
axs[0, 1].plot(x, y2)
axs[1, 0].plot(x, np.tan(x))
axs[1, 1].plot(x, np.arctan(x))

fig.subplots_adjust(hspace=0.5, wspace=0.5)
plt.show()
```


### Ejercicio 6: 

Reproduce la figura siguiente empleando sólo el estilo MATLAB (plt.\<funcion>\) y creando solo el contenido de los dos primeros gráficos (fila 1). Luego usa el estilo con métodos (objetos `axes`) para añadir el tercer gráfico.

```python
plt.figure(figsize=(9,5))
plt.subplot(2, 2, 1)
plt.plot(x, np.cos(x), label = "coseno")


plt.subplot(2, 2, 1)
plt.plot(x, np.sqrt(x), label = "raiz cuadrada")

plt.legend()
plt.title('Raíz Cuadrada y coseno')

# Acceso a Axes para modificar detalles
ax = plt.subplot(2, 2, 3)
ax.plot(x, np.sin(x))
ax.set_title('Seno')

plt.tight_layout()
plt.show()
```
