```python

import matplotlib.pyplot as plt
import numpy as np
```

### Ejercicio 1

Construye la siguiente figura con dos subplots utilizando el estilo MATLAB (es decir plt.\<funcion\>, no usando objetos `Axes`)

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)


plt.subplot(1, 2, 1)  # 1 fila, 2 columnas, primer subplot
plt.plot(x, y1)
plt.title('Seno')

plt.subplot(1, 2, 2)  # 1 fila, 2 columnas, segundo subplot
plt.plot(x, y2)
plt.title('Coseno')

plt.tight_layout()
plt.show()
```

### Ejercicio 2

Reproduce la imagen de la figura usando tres subplots en una columna, utilizando objetos `Axes`. Las dos primeras son las funciones anteriores y la tercera es la tangente (np.tan)

```python

fig, axs = plt.subplots(3, 1)  # 3 filas, 1 columna

axs[0].plot(x, y1, 'r')  # Rojo
axs[1].plot(x, y2, 'g')  # Verde
axs[2].plot(x, np.tan(x), 'b')  # Azul

plt.tight_layout()
plt.show()
```


### Ejercicio 3: Ajuste de Colores y Estilos de Lnea  

Modifica la primera figura para que se muestre de la siguiente manera

```python

fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.plot(x, np.sin(2 * x), 'm--')  # Magenta, lnea discontinua
ax2.plot(x, np.cos(2 * x), 'c-.')  # Cian, lnea-punto

plt.tight_layout()
plt.show()
```

### Ejercicio 4

Reproduce la figura siguiente. Recuerda que las funciones seno hiperblico y coseno hiperblico se obtienen as:
* np.sinh, seno hiperblico;
* np.cosh, coseno hiperblico;


```python
plt.subplot(2, 1, 1)
plt.plot(x, np.sinh(x), 'r')
plt.title('Seno Hiperblico')
plt.xlabel('x')
plt.ylabel('sinh(x)')

plt.subplot(2, 1, 2)
plt.plot(x, np.cosh(x), 'b')
plt.title('Coseno Hiperblico')
plt.xlabel('x')
plt.ylabel('cosh(x)')

plt.tight_layout()
plt.show()
```

### Ejercicio 5: Ajustando el Espaciado de Subplots

Reproduce la siguiente figura con cuatro subplots, ajusta el espaciado entre los subplots para evitar la superosicin (tal y como ves en la figura), emplea `subplots_adjust` y los parmetros hspace y wspace.

```python

fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(x, y1)
axs[0, 1].plot(x, y2)
axs[1, 0].plot(x, np.tan(x))
axs[1, 1].plot(x, np.arctan(x))

fig.subplots_adjust(hspace=0.5, wspace=0.5)
plt.show()
```
