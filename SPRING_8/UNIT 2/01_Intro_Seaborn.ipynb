### Consideraciones Generales en Visualizacin

* La visualizacin para el **anlisis NO es** la visualizacin para **presentar el resultado** del anlisis

Esto es lo primero que te tiene que quedar claro, y por qu es as? 

+ Primero, porque **TIENEN OBJETIVOS DISTINTOS**. El anlisis nos sirve para extraer informacin importante, respuestas a nuestras preguntas. La presentacin de resultados tiene como objetivo como mostrar esa informacin de la forma ms relevante para el destinatrario de esos resultados. Ojo, si el destinatario eres t mismo... O no hace falta que hagas una presentacin a ti mismo, o entonces t no eres el destinatario del resultado :-). 

Segundo, porque generalmente **TIENEN PBLICOS DISTINTOS**. El anlisis es para ti o para un grupo de colaboradores (el anlisis no es para tus jefes, salvo que sea colaborador en el proyecto, si tienes que mostrar el anlisis a jefes consideralo una Presentacin aunque en esa s puedas/debas ensear mucho del cmo). La presentacin es para el que ha encargado el proyecto, para convencer a un tercero, para informar, para quedar t como el gran sultn de los datos (es decir para un conjunto de "espectadores"), etc, etc.

De lo anterior surgen dos primeras guas importantes para  las visualizaciones:

+ En una nalisis tus visualizaciones pueden contener todos los datos comparados que tu cerebro pueda admitir. El lmite lo pones t o el grupo de trabajo.

+ En una presentacin la cantidad de informacin de la visualizacin depende del objetivo de la presentacin. Si:

    * Es una **presentacin de resultados y conclusiones**: La visualizacin debe contener **uno o como mucho dos DATOS destacados**, el resto aparecer como complemento esttico o comparativo (no importa el dato en s sino lo que refuerza o debilita a los anteriores), **por mensaje**.     * Es una **exposicin** de los mismos, por ejemplo **una infografa**: Entonces debe contener la mayor cantidad de **datos relevantes**, expuestos de la forma ms **clara, significativa y separada** posible.

### Seaborn y Matpotlib

**Seaborn** es una biblioteca de visualizacin de datos en Python que est construida sobre Matplotlib. Se puede considerar como una extensin o un "wrapper" de Matplotlib que proporciona una interfaz de alto nivel para dibujar grficos estadsticos atractivos y ms informativos. Como matplotlib y pandas, con los que se lleva de maravilla, tiene su propio alias (que ya conoces). Empecemos las importaciones:

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

Y cargamos unos datos para pintar:

```python
df = pd.read_csv("./data/titanic.csv")
```

* Matplotlib es muy potente y fexible, pero Seaborn tiene una interfaz ms accesible y estilos por defecto que son visualmente ms atractivos. Seaborn integra de manera ms profunda las estructuras de datos de Pandas y agrega capacidades de visualizacin estadstica. Esto incluye funciones para crear grficos complejos con menos cdigo, como grficos de distribucin, matrices de correlacin y muchos ms. Por ejemplo, generar un grfico de violn o un mapa de calor es ms sencillo y directo con Seaborn.

* Adems Seaborn naci con Pandas en la cabeza y facilita la visualizacin directa de DataFrames de Pandas, lo que es ideal para el anlisis de datos.

Entonces si vamos a pasarnos a Seaborn, para qu una unidad entera de Matplotlib?, te preguntars. Pues porque vamos a usar los dos conjuntamente:

* Usaremos Matplotlib para crear los subplots y ajustar los grficos creados con Seaborn, ya que las figuras de Seaborn son, en el fondo, figuras de Matplotlib. Y luego usaremos los mtodos de Matplotlib para personalizar los grficos de Seaborn, como modificar los ejes, aadir ttulos, etiquetas, y mucho ms, presentar varias graficas en una sola

Veamos dos ejemplos de como los usaremos (y de cmo los hemos estado usando ya):

#### Ejemplo 1: *Pinta Seaborn, decora Matplotlib*

```python
#hacemos un grafico diagrama de cajas de las columnas who y fare del DF del titanic
sns.boxplot(x="who", y="fare", data=df)

# usamos matplotlib para adornos
plt.title("Comparacion de Who con Fare en el dataset del Titanic")
plt.xlabel("Who")
plt.ylabel("fare")

```

#### Ejemplo 2: *Configura Matplotlib, pinta Seaborn*

```python
# configuramos la figura con matplotlib
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))# matriz de 2 filas x 2 columnas figuras con un tamao 10,8, que nos devolvera un objeto fig y un arry de objetos 
# axes que va ir al valor axs

# Primer subplot con un grfico de Seaborn
sns.boxplot(x='who', y='fare', data=df, ax=axs[0, 0])# axs(0,0) decimos el indice donde ira esa imagen

# Segundo subplot con otro grfico de Seaborn
sns.scatterplot(x='age', y='fare', data=df, ax=axs[0, 1])# grafico de dispersion

# Tercer subplot
sns.histplot(df['age'], kde=True, ax=axs[1, 0])# distribucion de edad

# Cuarto subplot
sns.violinplot(x='embark_town', y='age', data=df, ax=axs[1, 1]);

axs[0,0].set_title("who vs fare")
axs[0,1].set_title("age vs fare")
axs[1,0].set_title("Distrib. Age")
axs[1,1].set_title("Embarked_town vs Age, violins")

plt.tight_layout()

```
