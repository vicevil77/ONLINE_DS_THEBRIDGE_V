### Distribucin: Histrogramas, Funciones Densidad (KDE), Combinado


Empezamos con el histograma, ese viejo compaero ya (tanto que no te hago escribir)

```python
fig,axs = plt.subplots(1,1,figsize= (8,4))
sns.histplot(df_viajes["ingresos"]/1000,
             kde=False,#que no nos pinte las 2 cosas, el histograma y la densidad de probtabilidad
             color='r',
             bins=100, ax = axs) # Otra forma, sin recurrir al dataframe como argumento
axs.set_xlabel("Ingresos  (millones)")
axs.set_ylabel("Num_Vuelos");
```

Veamos ahora el Kernel Density Estimation (KDE) o nuestra estimacin de la funcin de densidad:

```python
#AC
fg = sns.displot(x="customer_lifetime_value", data = df_seguros, kind="kde", height=4, aspect=3)
# height controla la altura del grfico
# Aspect la relacin ente ancho/alto( 3 veces el ancho sobre el alto)
#kind= "kde"(auqnue le podria poner tb que dibujara el histograma)
# Ojo este grfico se puede asignar a un "axes" pero en realidad es para mostrar uno por "figure" (es slo por completitud)
# si yo quiero calcular la probabilidad CLV entre 20000 y 40000, tendria que calcular lo que queda debajo de la curva entre esos intervalos
```

Y ahora los dos juntos:

```python
fig,axs = plt.subplots(1,1,figsize= (8,4))
sns.histplot(df_viajes["ingresos"]/1000,
             kde=True,#nos pinte las 2 cosas, el histograma y la densidad de probtabilidad
             color='r',
             bins=100, ax = axs) # Otra forma, sin recurrir al dataframe como argumento
axs.set_xlabel("Ingresos  (millones)")
axs.set_ylabel("Num_Vuelos");
```

### Enjambre 

Este es nuevo y es interesante para anlisis, no s hasta que punto lo es para visualizacin salvo que en l se vea algo que no se pueda mostrar de otra forma. Este diagrama nos muestra los puntos.
lo que hace este diagrama es coger una variable continua la considera como si fuera una variable ordinal, es decir tantos valores como categorias pueda adquirir la funcion y para cada uno de ellos, pinta o dibuja o cuenta un punto para cada valor, pudiendo ver la dispersion

```python
plt.figure(figsize=(12,6))# si yo creo en la misma celda una figura , el siguiente grafico que se cxree se asigna a esta figura, por eso la relacion de matplotlib con seaborn
sns.swarmplot(x ="consumo_kg", data=df_viajes, color="black");
```

Nos muestra para los valores de consumo los puntos sin solape en cada valor. Se suele **complementar con un grfico de caja o de violn para ver la dispersin y la distribucin:**

```python
plt.figure(figsize=(8,6))
sns.boxplot(x = "consumo_kg", data = df_viajes)
sns.swarmplot(x = "consumo_kg", data=df_viajes,
             color="black");# como es de la mimsa variable el mimso seaborn superpone las imagenes sin necesidad de tocar los ejes
```

Y para terminar esta parte, los violines (musica maestro): ( el violin es mucho mas visual que el boxplot)

```python
#AC# ya que no tengo la ultima version de seaborn(no me intersa ya que entra en conflicto con ydate), tengo que poner xlabel e ylabel, poner vert para lka orientacion y sin color
plt.figure(figsize=(20,5))
plt.violinplot(df_viajes["consumo_kg"], vert= False)
plt.xlabel('Consumo (kg)')
plt.title('Distribucin del Consumo')
plt.show()

# la estimacion de la densidad por arriba por lo que nos da un poco la distribucion de valores, por debajo un mini boxplot, la mediana y otro boxplot
```
